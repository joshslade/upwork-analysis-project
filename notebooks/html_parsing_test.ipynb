```python
import re
import ast
from itertools import islice


def extract_final_args_block(js):
    # Find the position of the last '));'
    end = js.rfind('));')
    if end == -1:
        return None

    # Work backwards to find the matching '('
    pos = end - 1
    depth = 0
    while pos >= 0:
        if js[pos] == ')':
            depth += 1
        elif js[pos] == '(':
            if depth == 0:
                # Found the matching opening parenthesis
                return js[pos+1:end]  # Exclude the outer parentheses
            else:
                depth -= 1
        pos -= 1
    return None

# Read the script content from file
with open("../data/nuxt_script_content.js", "r", encoding="utf-8") as f:
    js = f.read()

# 1. Extract parameter names from the function definition
param_match = re.search(r'function\s*\((.*?)\)', js, re.DOTALL)
params = [p.strip() for p in param_match.group(1).split(',')] if param_match else []

# 2. Extract the argument list from the function call at the end
args_str = extract_final_args_block(js)

# 3. Split the arguments, handling quoted strings and special JS values
# Replace JS-specific values with Python equivalents
args_str_py = (
    args_str
    .replace('void 0', 'None')
    .replace('null', 'None')
    .replace('true', 'True')
    .replace('false', 'False')
    .replace('\\u002F', '/')
)

# Use ast.literal_eval after wrapping in brackets to parse as a Python list
try:
    args_list = ast.literal_eval(f"[{args_str_py}]")
except Exception as e:
    print("Error parsing arguments:", e)
    args_list = []

# 4. Zip parameters and values
param_value_dict = dict(zip(params, args_list))

# Example: print first 10 pairs from the dictionary
for name, value in islice(param_value_dict.items(), 10):
    print(f"{name}: {value}")

# param_value_dict now contains all parameter-value mappings
```

```python
import re
import json

# Find 'jobs:' followed by whitespace and a '['
jobs_match = re.search(r'jobs\s*:\s*\[', js)
if not jobs_match:
    print("No jobs array found.")


start = jobs_match.end() - 1  # position at the '['

# Now, find the matching closing brace for this object
depth = 0
for i in range(start, len(js)):
    if js[i] == '[':
        depth += 1
    elif js[i] == ']':
        depth -= 1
        if depth == 0:
            end = i + 1
            break

obj_str = js[start:end]

# Clean up JS to JSON (basic, may need more for complex cases)
obj_str = obj_str.replace("'", '"')
obj_str = re.sub(r'(?<!["\\])(\b\w+\b)\s*:', r'"\1":', obj_str)
obj_str = obj_str.replace('undefined', 'null')
obj_str = obj_str.replace('void 0', 'null')
obj_str = obj_str.replace('false', 'false')
obj_str = obj_str.replace('true', 'true')
obj_str = obj_str.replace('\u002F', '/')

# Try to load as JSON
try:
    obj = json.loads(obj_str)
except Exception as e:
    print("Error parsing JSON:", e)
    with open("return_object_raw.json", "w", encoding="utf-8") as f:
        f.write(obj_str)
```
